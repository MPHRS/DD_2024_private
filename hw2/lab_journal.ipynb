Решение с использованием Git LFS:

    Установите Git LFS

    Для начала, вам нужно установить Git LFS на вашу машину. Если он еще не установлен, выполните следующую команду:

    Для Linux (Ubuntu/Debian):

sudo apt-get install git-lfs

Для macOS:

brew install git-lfs

Для Windows: Следуйте инструкциям на официальной странице Git LFS.

Инициализация Git LFS в вашем репозитории

После установки Git LFS, вам нужно инициализировать его в вашем репозитории:

git lfs install

Отслеживание больших файлов с помощью Git LFS

Чтобы Git LFS отслеживал файлы, которые превышают лимит в 100 MB (например, ваши .fastq.gz файлы), выполните команду:

git lfs track "*.fastq.gz"

Это скажет Git LFS отслеживать все файлы с расширением .fastq.gz в вашем репозитории.

Добавьте изменения в индекс Git и сделайте коммит

После того как вы настроили отслеживание больших файлов, добавьте изменения в индекс и сделайте новый коммит:

git add .gitattributes
git add hw2/spades/corrected/SRR292678sub_S1_L001_R1_00100.0_0.cor.fastq.gz
git add hw2/spades/corrected/SRR292678sub_S1_L001_R2_00100.0_0.cor.fastq.gz
git commit -m "Добавлены большие файлы с использованием Git LFS"
